```python
from scipy import stats
import pandas as pd
import plotly.express as plt
import numpy as np
import os
import statsmodels.api as sm
from matplotlib import pyplot
```

```python
property = os.path.join('property_sales.csv')
property_df = pd.read_csv(property)
property_df.head(15)
```

```python
population = property_df['Downpayment'].dropna()
```

```python
y = property_df['Gross_Sale']
```

```python
sample = np.random.choice(population, size=1000)
```

```python
stats.ttest_1samp(sample, y.mean())
```
